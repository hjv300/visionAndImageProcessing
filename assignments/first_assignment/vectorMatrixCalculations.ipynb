```python
def matrixAsso(vector):
    matrix = np.zeros((3, 3))
    matrix[0][1] = -1*vector[2]
    matrix[0][2] = vector[1]
    matrix[1][0] = vector[2]
    matrix[1][2] = -1*vector[0]
    matrix[2][0] = -1*vector[1]
    matrix[2][1] = vector[0]
    return matrix

print(np.dot(matrixAsso(t), t))
print("\n----------\n")
print(np.dot(matrixAsso(t), a))
```

What do you get?
- In the first one I get a matrix of Zeros, which corresponds to the properties of the skew-simmetric matrix ($\^{t}^T = -t | \^{t} \cdot t = [0, 0, 0]$).
- the second one is a one dimension matrix with the shown results.

For a 3D vector v, the operation v -> t x v is linear. So it can be represented by a matrix. Which one ?
- the skew-symmetric matrix would satisfy the representation.

```python
assert np.all(matrixAsso(t).T == -matrixAsso(t))
print('the assertion is true! hurray!')
```

<center> 3.1. Compute the derivative:</center>

$$f(x) = e^{-\frac{x^{2}}{2}}\\
f'(x) = e^{-\frac{x^{2}}{2}} \cdot \frac{d}{dx}(-\frac{x^{2}}{2})\\ 
f'(x) = e^{-\frac{x^{2}}{2}} \cdot (-x)\\
f'(x) = -xe^{-\frac{x^{2}}{2}}$$

<center> 3.2. Compute the derivative of the derivative:</center>

$$f'(x) = -xe^{-\frac{x^{2}}{2}}\\
f''(x) = (-x)'\cdot e^{-\frac{x^{2}}{2}} + (-x)\cdot (e^{-\frac{x^{2}}{2}})'\\
f''(x) = -e^{-\frac{x^{2}}{2}} - x \cdot (-xe^{-\frac{x^{2}}{2}})\\
f''(x) = -e^{-\frac{x^{2}}{2}} + x^2\cdot e^{-\frac{x^{2}}{2}}\\
f''(x) = e^{-\frac{x^{2}}{2}} \cdot (x^2 - 1)$$

<center> 3.3. Plot on the same graph f(x), f'(x) and f''(x), say for x âˆˆ [-7, 7] :</center>

```python
import matplotlib.pyplot as plt
import numpy as np
from math import e

x_values = np.linspace(-7, 7)  
plt.figure()

f1 = [e**(-x**2 / 2) for x in x_values]
f2 = [-x * e**(-x**2 / 2) for x in x_values]
f3 = [e**(-x**2 / 2) * (x**2 - 1) for x in x_values]

plt.plot(x_values, f1, label='e**(-x**2/2)')
plt.plot(x_values, f2, label='-x * e**(-x**2/2)')
plt.plot(x_values, f3, label='e**(-x**2/2) * (x**2 - 1)')

plt.legend() 

plt.show()
```

<center> 3.8. Compute the derivative of the derivative with respect to x:</center>

$$f'(x, t) = -\frac{x}{t\sqrt{2\pi t}}e^{-\frac{x^{2}}{2t}}\\
f''(x, t) = (-\frac{x}{t\sqrt{2\pi t}})'\cdot e^{-\frac{x^2}{2t}} + (-\frac{x}{t\sqrt{2\pi t}})\cdot(e^{-\frac{x^2}{2t}})'\\
f''(x, t) = (-\frac{1}{t\sqrt{2\pi t}})\cdot e^{-\frac{x^2}{2t}} + (-\frac{x}{t\sqrt{2\pi t}})\cdot (-\frac{x}{t^2}e^{-\frac{x^2}{2t}})\\
f''(x, t) = -\frac{1}{t^2\sqrt{2\pi t}} e^{-\frac{x^2}{t}} + \frac{x^2}{t^3\sqrt{2\pi t}} e^{-\frac{x^2}{2t}}\\$$


<center>Then show that:</center>

$$\frac{\partial f}{\partial t}(x, t)=\frac{1}{2} \frac{\partial^2 f}{\partial x^2}(x, t)\\
f'(x, t) = -\frac{1}{2}\frac{1}{\sqrt{2\pi t^3}}e^{-\frac{x^2}{2t}} + \frac{1}{\sqrt{2\pi t}}e^{-\frac{x^2}{2t}}(\frac{x^2}{2t^2}) \neq f''(x, t) = \frac{1}{2}(-\frac{1}{t^2\sqrt{2\pi t}} e^{-\frac{x^2}{t}} + \frac{x^2}{t^3\sqrt{2\pi t}} e^{-\frac{x^2}{2t}})
$$


<center>Plot these functions for t = 1, 2, 4 with x âˆˆ [-15, 15]</center>

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Define the functions
def f_prime(x, t):
    term1 = -0.5 * (1 / (math.sqrt(2 * math.pi * t**3))) * math.exp(-x**2 / (2 * t))
    term2 = (1 / (math.sqrt(2 * math.pi * t))) * math.exp(-x**2 / (2 * t)) * (x**2 / (2 * t**2))
    return term1 + term2
def f_double_prime(x, t):
    term1 = -0.5 * (1 / (t**2 * math.sqrt(2 * math.pi * t))) * math.exp(-x**2 / t)
    term2 = 0.5 * (x**2 / (t**3 * math.sqrt(2 * math.pi * t))) * math.exp(-x**2 / (2 * t))
    return term1 + term2

x_values = np.linspace(-15, 15, 400)

plt.figure()
#Plot the functions
for t in [1, 2, 4]:
    y_values_prime = [f_prime(x, t) for x in x_values]
    y_values_double_prime = [f_double_prime(x, t) for x in x_values]
    plt.plot(x_values, y_values_prime, label=f'f\'(x, t={t})')
    plt.plot(x_values, y_values_double_prime, label=f'f\'\'(x, t={t})')
# Add a legend and labels
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of f\'(x, t) and f\'\'(x, t) for t = 1, 2, 4')

plt.show()
```
